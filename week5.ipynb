```python
import numpy as np
import pandas as pd

# Fix a random seed so we have a predicable outcome
np.random.seed(0)

# Read in the data from a URL (local file in this case)
url="Phishing.csv"
df=pd.read_csv(url)

# Replace -1 with 0
df = df.replace(-1, 0)

df
```

```python
# From https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
# Split the set in a predicable way for test and training

X = df.iloc[:,:-1] # Features
y = df['Result']   # Target, we are using the features to try and get this result
from sklearn.model_selection import train_test_split
(X_train,X_test,y_train,y_test) = train_test_split(X,y, test_size=0.2, random_state=0)
X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping

# Build the model
model = Sequential()
model.add(Dense(40, input_dim = 30 ,activation='relu'))
model.add(Dense(30, input_dim = 30, activation='relu'))
model.add(Dense(1, input_dim = 30, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Define early stopping function
stop = EarlyStopping(monitor='val_loss', mode='min', verbose=0)

model.fit(X_train, y_train, epochs=128, batch_size=64, callbacks=[stop])

# evaluate the keras model
_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))

```
