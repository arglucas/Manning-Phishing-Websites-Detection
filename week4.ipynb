```python
import numpy as np
import pandas as pd

# Fix a random seed so we have a predicable outcome
np.random.seed(0)

# Read in the data from a URL (local file in this case)
url="Phishing.csv"
df=pd.read_csv(url)

# Replace -1 with 0
df = df.replace(-1, 0)

df
```

```python
df.iloc[:,:-1]
```

```python
df["Result"]
```

```python
# From https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
# Split the set in a predicable way for test and training

X = df.iloc[:,:-1] # Features
y = df['Result']   # Target, we are using the features to try and get this result
from sklearn.model_selection import train_test_split
(X_train,X_test,y_train,y_test) = train_test_split(X,y, test_size=0.2, random_state=0)
X_train
```

```python
# Train a logistic regression model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
y_pred
```

```python
from sklearn import metrics
# How accurate our model was at predicting the correct result
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# How often we were correct when we made a guess, ie. when we thought it was a good email how often was it
print("Precision:",metrics.precision_score(y_test, y_pred))
# How often we correctly identified something given the condition is in the test set
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
target_names = ['Phishing Websites', 'Normal Websites']
print(metrics.classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = dict(
    penalty = ['l1', 'l2'],
    C = [0.8, 0.9, 1.0],
    tol = [0.01, 0.001, 0.0001],
    max_iter = [100, 150, 200, 250]
)

logreg2 = LogisticRegression(solver='saga')
rslr2 = RandomizedSearchCV(logreg2, params, random_state=0)
search = rslr2.fit(X,y)
search
```
