```python
# Adjusted from: https://datascience.stackexchange.com/questions/10459/calculation-and-visualization-of-correlation-matrix-with-pandas
import matplotlib.cm as cm
def correlation_matrix(df, labels=None, show_all_labels=True):
    if labels == None:
        labels = df.columns
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 30)
    cax = ax1.imshow(df.corr(), interpolation="nearest", cmap=cmap)
    ax1.grid(True)
    if show_all_labels:
        ax1.set_xticks(np.arange(len(labels)))
        ax1.set_yticks(np.arange(len(labels)))
    ax1.set_xticklabels(labels,fontsize=8)
    ax1.set_yticklabels(labels,fontsize=8)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=np.arange(-1,1.05, step=0.05))
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Allows me to see where features are correlated, there is the obvious correlation between identical items
# (diagonal top-left to bottom right) but there are other correlations around right-clicks, onmouseover etc which
# make some logical sense.
plt.rcParams["figure.figsize"] = (12,9) # Size in inches
correlation_matrix(df)
```
